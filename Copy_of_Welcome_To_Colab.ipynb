## **Cleaning Data**

**Relevant Data of External Stock Data**

```
import pandas as pd

countries = ['Angola', 'Botswana', 'Congo, Dem. Rep.', 'Lesotho', 'Malawi', 'Mozambique', 'South Africa', 'Tanzania', 'Zambia', 'Zimbabwe']
filtered_df = df[df['Country Name'].isin(countries)]

# Remove duplicates from filtered_df
filtered_df.drop_duplicates(inplace=True)

# Find rows with missing values in any column for the selected countries
rows_with_missing_values = filtered_df[filtered_df.isnull().any(axis=1)]

# Get the years where any of the selected countries have missing data
years_with_missing_data = rows_with_missing_values['Time'].unique()

# Filter the DataFrame to exclude rows with missing data for the specified years
consistent_df = filtered_df[~filtered_df['Time'].isin(years_with_missing_data)]

grouped_df = consistent_df.groupby('Country Name')

for country, group in grouped_df:
    print(f"Country: {country}")
    print(group.to_string)
    print("\n")
```

**Relevant Data Of Debt Of GDP**

```
import pandas as pd

countries = ['Angola', 'Botswana', 'Congo, Dem. Rep.', 'Lesotho', 'Malawi', 'Mozambique', 'South Africa', 'Tanzania', 'Zambia', 'Zimbabwe']

# Filter by countries
filtered_df2 = df2[df2['Country Name'].isin(countries)]

# Get the last column name
last_column_name = df2.columns[-1]

# Select only the 'Country Name' and the last column
filtered_data = filtered_df2[['Country Name', last_column_name]]

# Group by 'Country Name'
grouped_df2 = filtered_data.groupby('Country Name')

for country, group in grouped_df2:
    print(f"Country: {country}")
    print(group[last_column_name])  # Print data from the last column for the country
    print("\n")
```

```
countries = ['Angola', 'Botswana', 'Congo, Dem. Rep.', 'Lesotho', 'Malawi', 'Mozambique', 'South Africa', 'Tanzania', 'Zambia', 'Zimbabwe']
filtered_df3 = df3[df3['Country'].isin(countries)]
grouped_df3 = filtered_df3.groupby('Country')

for country, group in grouped_df3:
    print(f"Country: {country}")
    print(group.head())
    print("\n")
```

**Exploratory Data Analysis**

```
import matplotlib.pyplot as plt

# Assuming 'filtered_data' contains your data as described
countries = filtered_data['Country Name'].tolist()
debt_gdp = filtered_data[last_column_name].tolist()

plt.figure(figsize=(12, 6))  # Adjust figure size as needed
plt.bar(countries, debt_gdp, color='blue')
plt.title('Country Versus Debt of GDP in 2015')
plt.xlabel('Country Names')
plt.ylabel('Debt (% of GDP)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
