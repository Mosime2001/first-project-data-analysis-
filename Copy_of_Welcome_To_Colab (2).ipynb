```
import pandas as pd

def calculate_stats(file_path):
    """
    Calculates the median, mean, mode, standard deviation, and range
    for 'External debt stocks In US dollars' for Botswana and Mozambique.
    """
    try:
        df = pd.read_excel(file_path)
    except FileNotFoundError:
        return "File not found. Please ensure the file path is correct."

    # Filter data for Botswana and Mozambique
    countries = ['BWA', 'MOZ']
    filtered_df = df[df['Country Code'].isin(countries)].copy()

    # Convert 'External debt stocks In US dollars' to numeric
    filtered_df.loc[:, 'External debt stocks In US dollars'] = pd.to_numeric(filtered_df['External debt stocks In US dollars'], errors='coerce')

    # Drop rows where 'External debt stocks In US dollars' is NaN after conversion
    filtered_df.dropna(subset=['External debt stocks In US dollars'], inplace=True)

    # Group by country and calculate statistics
    grouped = filtered_df.groupby('Country Code')['External debt stocks In US dollars']

    results = {}
    for country, data in grouped:
        results[country] = {
            'mean': data.mean(),
            'median': data.median(),
            'mode': data.mode().tolist(),  # Mode can have multiple values
            'std': data.std(),
            'range': data.max() - data.min()
        }

    return results

# Example usage:
file_path = 'P_Data.xlsx'  # Replace with your file path
stats = calculate_stats(file_path)

if isinstance(stats, str):
    print(stats)
else:
    for country, data in stats.items():
        print(f"Country: {country}")
        for stat, value in data.items():
            print(f"  {stat}: {value}")
        print("\n")
```

```
import pandas as pd

def calculate_average_gdp(csv_file, countries, start_year, end_year, year_2022=2022):
    """
    Calculates the average GDP for specified countries over a given period and retrieves GDP for 2022.

    Parameters:
    - csv_file (str): Path to the CSV file containing GDP data.
    - countries (list): A list of country names to calculate the average GDP for.
    - start_year (int): The starting year for calculating the average GDP.
    - end_year (int): The ending year for calculating the average GDP.

    Returns:
    - dict: A dictionary where keys are country names and values are dictionaries containing:
            - 'average_gdp': The average GDP over the specified period.
            - 'gdp_2022': The GDP value for the year 2022.
            Returns None if the CSV file is not found.
    """

    try:
        # Read the CSV file into a pandas DataFrame
        df = pd.read_csv(csv_file)
        df = df.set_index('Country')
    except FileNotFoundError:
        print(f"Error: The file {csv_file} was not found.")
        return None

    results = {}

    for country in countries:
        if country not in df.index:
            print(f"Warning: {country} not found in the dataset.")
            continue

        # Extract GDP data for the specified period
        years = [str(year) for year in range(start_year, end_year + 1)]
        country_data = df.loc[country, years].astype(float)

        # Calculate the average GDP
        average_gdp = country_data.mean()

        # Get the GDP for 2022
        gdp_2022 = df.loc[country, str(year_2022)]

        # Store the results
        results[country] = {
            'average_gdp': average_gdp,
            'gdp_2022': gdp_2022
        }

    return results

# Example usage:
csv_file = 'GDP.csv'
countries = ['Botswana', 'Mozambique']
start_year = 2005
end_year = 2015

results = calculate_average_gdp(csv_file, countries, start_year, end_year)

if results:
    for country, data in results.items():
        print(f"Country: {country}")
        print(f"Average GDP ({start_year}-{end_year}): {data['average_gdp']:,.2f}")
        print(f"GDP (2022): {data['gdp_2022']}")
        print("-" * 30)
```

```
import pandas as pd
import matplotlib.pyplot as plt

# Load your data (replace 'P_Data.xlsx' with your file path)
df = pd.read_excel('P_Data.xlsx')

# Filter data for Botswana and Mozambique
filtered_df = df[df['Country Name'].isin(['Botswana', 'Mozambique'])]

# Clean 'External debt stocks In US dollars' column
filtered_df['External debt stocks In US dollars'] = pd.to_numeric(filtered_df['External debt stocks In US dollars'], errors='coerce')
filtered_df['External debt stocks In US dollars'].fillna(0, inplace=True)

# Remove rows with empty 'Counterpart-Area Name'
filtered_df = filtered_df[filtered_df['Counterpart-Area Name'] != ' ']  # Assuming empty areas are represented by a single space

# Get top 8 counterpart areas for Botswana and Mozambique separately
top_8_botswana = filtered_df[filtered_df['Country Name'] == 'Botswana'].groupby('Counterpart-Area Name')['External debt stocks In US dollars'].sum().nlargest(8).index.tolist()
top_8_mozambique = filtered_df[filtered_df['Country Name'] == 'Mozambique'].groupby('Counterpart-Area Name')['External debt stocks In US dollars'].sum().nlargest(8).index.tolist()

# Combine the top 8 for both countries
top_16_counterparts = top_8_botswana + top_8_mozambique

# Filter data for top 16 counterpart areas
filtered_data = filtered_df[filtered_df['Counterpart-Area Name'].isin(top_16_counterparts)]

# Group and aggregate data
grouped_data = filtered_data.groupby(['Country Name', 'Counterpart-Area Name'])['External debt stocks In US dollars'].sum().reset_index()

# Pivot data for grouped bar chart
pivoted_data = grouped_data.pivot(index='Counterpart-Area Name', columns='Country Name', values='External debt stocks In US dollars')

# Create grouped bar chart
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
pivoted_data.plot(kind='bar', figsize=(10, 6))
plt.title('External Debt Stock Distribution for Top 8 Counterpart Areas (Botswana & Mozambique)')
plt.xlabel('Counterpart Area')
plt.ylabel('Total External Debt Stock (in US dollars)')
plt.legend(title='Country')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load Debt-to-GDP data
debt_gdp_df = pd.read_excel('DEBT- of GDP.xlsx')
debt_gdp_df = debt_gdp_df[debt_gdp_df['Country Name'].isin(['Botswana', 'Mozambique'])]
debt_gdp_df = debt_gdp_df.set_index('Country Name')

# Transpose the DataFrame to have years as index and countries as columns
debt_gdp_df = debt_gdp_df.T

# Convert index to datetime objects (adjust format if needed)
debt_gdp_df.index = pd.to_datetime(debt_gdp_df.index, errors='coerce')  # Use format='%Y' if years are in 'YYYY' format

# Remove rows with NaT (Not a Time) in the index
debt_gdp_df = debt_gdp_df[debt_gdp_df.index.notnull()]

# Create plot
plt.figure(figsize=(12, 6))

for country in ['Botswana', 'Mozambique']:
    debt_levels = debt_gdp_df[country].values  # Get debt levels for the country

    # Calculate metrics
    rolling_std = pd.Series(debt_levels).rolling(window=5).std()  # 5-year rolling std
    mean_debt = pd.Series(debt_levels).rolling(window=5).mean()  # 5-year rolling mean
    min_max_range = pd.Series(debt_levels).rolling(window=5).agg([np.min, np.max])  # 5-year rolling min/max

    # Standard deviation band
    plt.fill_between(
        debt_gdp_df.index,
        mean_debt - rolling_std,
        mean_debt + rolling_std,
        alpha=0.3,
        label=f'{country} Â±1 Std Dev',
    )

    # Min-max range markers
    plt.scatter(
        debt_gdp_df.index,
        min_max_range['min'],  # Corrected column name
        marker='_',
        s=100,
        label=f'{country} Rolling Min',
    )
    plt.scatter(
        debt_gdp_df.index,
        min_max_range['max'],  # Corrected column name
        marker='_',
        s=100,
        label=f'{country} Rolling Max',
    )

    # Mean trend line
    plt.plot(debt_gdp_df.index, mean_debt, lw=2, label=f'{country} 5-Year Mean')

plt.title('Debt Volatility Analysis: Stability and Risk Profile (Botswana & Mozambique)', pad=15)
plt.xlabel('Time')
plt.ylabel('Debt-to-GDP Ratio (%)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside the plot
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```
